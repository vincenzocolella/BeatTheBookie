```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = 'https://www.betfair.com/sport/football'
response = requests.get(url)
html_content = response.content

soup = BeautifulSoup(html_content, 'html.parser')
match_data = []

# find all the match containers on the page
match_containers = soup.find_all('div', {'class': 'module-placeholder'})

soup = BeautifulSoup(html_content, 'html.parser')
match_data = []

avg_odds_home_win = home_odds = match_containers[0].find('span', {'class': 'ui-runner-price ui-924_367601714-13288774 ui-display-decimal-price ui-price-up'}).text.strip()
print(avg_odds_home_win)
# loop through each match container and extract the match data elements
for match_container in match_containers:
    match_id = match_container.get('data-market-id')
    league = match_container.find('div', {'class': 'coupon-title'}).text.strip()
    match_date = match_container.find('span', {'class': 'start-time'}).text.strip()
    home_team = match_container.find('span', {'class': 'home-team-name'}).text.strip()
    home_score = match_container.find('span', {'class': 'home-team-score'}).text.strip()
    away_team = match_container.find('span', {'class': 'away-team-name'}).text.strip()
    away_score = match_container.find('span', {'class': 'away-team-score'}).text.strip()
    avg_odds_home_win = home_odds = match_container.find('span', {'class': 'ui-runner-price ui-924_367601714-13288774 ui-display-decimal-price ui-price-up'}).text.strip()
    avg_odds_draw = match_container.find('span', {'class': 'odds draw-back selection-1'}).text.strip()
    avg_odds_away_win = match_container.find('span', {'class': 'odds away-back selection-2'}).text.strip()
    max_odds_home_win = match_container.find('span', {'class': 'odds home-back selection-0'}).get('data-oc-price')
    max_odds_draw = match_container.find('span', {'class': 'odds draw-back selection-1'}).get('data-oc-price')
    max_odds_away_win = match_container.find('span', {'class': 'odds away-back selection-2'}).get('data-oc-price')
    top_bookie_home_win = match_container.find('span', {'class': 'odds home-back selection-0'}).get('data-oc-oc-name')
    top_bookie_draw = match_container.find('span', {'class': 'odds draw-back selection-1'}).get('data-oc-oc-name')
    top_bookie_away_win = match_container.find('span', {'class': 'odds away-back selection-2'}).get('data-oc-oc-name')
    n_odds_home_win = len(match_container.find_all('span', {'class': 'odds home-back selection-0'}))
    n_odds_draw = len(match_container.find_all('span', {'class': 'odds draw-back selection-1'}))
    n_odds_away_win = len(match_container.find_all('span', {'class': 'odds away-back selection-2'}))
    
    # append the match data elements to the match_data list
    match_data.append([match_id, league, match_date, home_team, home_score, away_team, away_score,
                       avg_odds_home_win, avg_odds_draw, avg_odds_away_win, max_odds_home_win,
                       max_odds_draw, max_odds_away_win, top_bookie_home_win, top_bookie_draw,
                       top_bookie_away_win, n_odds_home_win, n_odds_draw, n_odds_away_win])
    
    
    
match_df = pd.DataFrame(match_data, columns=['match_id', 'league', 'match_date', 'home_team', 'home_score',
                                              'away_team', 'away_score', 'avg_odds_home_win', 'avg_odds_draw',
                                              'avg_odds_away_win', 'max_odds_home_win', 'max_odds_draw',
                                              'max_odds_away_win', 'top_bookie_home_win', 'top_bookie_draw',
                                              'top_bookie_away_win', 'n_odds_home_win', 'n_odds_draw',
                                              'n_odds_away_win'])

match_df.to_csv('betfair_data.csv', index=False)
```
